```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# Define the population (a fair six-sided die)
population = np.array([1, 2, 3, 4, 5, 6])

# Define the number of samples to take
num_samples = 1000

# Define different sample sizes
sample_sizes = [1, 2, 5, 10, 50]

# Create a figure to plot the results
plt.figure(figsize=(12, 8))

# Iterate over each sample size
for i, sample_size in enumerate(sample_sizes, 1):
    # Generate sample means for the current sample size
    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]
    
    # Plot the histogram of sample means
    plt.subplot(2, 3, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7)
    plt.title(f'Sample Size: {sample_size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

# Add a title to the overall figure
plt.suptitle('Demonstration of Central Limit Theorem', fontsize=16)
plt.tight_layout()
plt.show()
```

```python

# Sample data
data = [10, 12, 14, 15, 16, 18, 20]

# Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data)

# Sample size
n = len(data)

# Confidence level (e.g., 95%)
confidence_level = 0.95

# Calculate standard error of the mean (SEM)
SEM = sample_std / np.sqrt(n)

# Calculate margin of error (MOE) using the standard normal distribution
z_score = norm.ppf((1 + confidence_level) / 2)  # Two-tailed test
MOE = z_score * SEM

# Calculate confidence interval (CI)
lower_limit = sample_mean - MOE
upper_limit = sample_mean + MOE

print("Sample Mean:", sample_mean)
print("Standard Error of the Mean (SEM):", SEM)
print("Margin of Error (MOE):", MOE)
print("Confidence Interval (CI):", (lower_limit, upper_limit))
```
